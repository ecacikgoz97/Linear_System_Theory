```python
import numpy as np
from numpy import linalg as la
import matplotlib.pyplot as plt
import random
```

## Randomly Create I_A and C

```python
# Random number generator
Total = np.random.randint(2, size=200)

I_A = []
I_B = []

# Creates I_A and I_B
cntr=1
for person in Total:
    if person == 0:
        I_A.append(cntr)
    else:
        I_B.append(cntr)
    cntr+=1
```

```python
# Convert to Array
I_A = np.array(I_A)
I_B = np.array(I_B)
```

```python
print(I_A)
I_A.shape
```

```python
print(I_B)
I_B.shape
```

```python
# Creating the C matrix
C = np.ones((200,200), dtype=int)
for i in range(200):
    for j in range(200):
        if ((i in I_A) and (j in I_B)) or ((i in I_B) and (j in I_A)):
            rand_num = random.randint(0,9)
            if (rand_num == 0):
                C[i,j] = 1
                C[j,i] = 1  
            else:
                C[i,j] = 0
                C[j,i] = 0 
        elif ((i in I_A) and (j in I_A)) or ((i in I_B) and (j in I_B)):
            rand_num = random.randint(0,9)
            if (rand_num == 0) or (rand_num == 1) or (rand_num == 2) or (rand_num == 3) or (rand_num == 4) or (rand_num == 5):
                C[i,j] = 1
                C[j,i] = 1 
            else:
                C[i,j] = 0
                C[j,i] = 0 
```

## Display C Using imshow()

```python
plt.figure(figsize = (20,20))
plt.imshow(C, 'Greys');
# We can think each pixels(Squares) as each person and the white pixels are the people who are in different 
# communities and the minory black pixels are the people in same communities.
```

```python
# C Matrix
print("C is a",C.shape," Matrix")
print(C)
```

## Finding Eigenvalues and Eigenvectors

```python
# Finding Eigenvalues and eigenvectors
Eig, Eigenvectors = la.eig(C)
```

```python
print("The Eigenvalues are :")
print(Eig)
```

```python
print("The Eigenvectors are :")
print(Eigenvectors)
```

```python
# Plots the maximum score Eigenvector
plt.figure(figsize = (10,10))
plt.stem(Eigenvectors[1]);
# It helps to find I_A in a way that: when the element is positive it is corresponds to class I_A. But We must 
# do some iterations to achieve the best score
```
